# File takes in .nii and returns yt ds

```python
def yt_load_nii(file):
    data = {}
    sub = 'sub'
    data[sub]=nib.load(file).get_fdata()
    print(data[sub].shape)
    bbox = np.array([[0,(data[sub].shape[0])],
                     [0, (data[sub].shape[1])],
                     [0, (data[sub].shape[2])]],dtype='f8')

    ds = yt.load_uniform_grid(data, domain_dimensions = ((data[sub].shape[0]),(data[sub].shape[1]),(data[sub].shape[2])),bbox = bbox,
                        geometry = ('cartesian'))
    return ds
```

```python
ds = yt_load_nii('data/Eye_brain/derivatives/MP2RAGEToolboxSPM/sub-01/ses-01/sub-01_ses-01_run-1_T1map.nii')
```
