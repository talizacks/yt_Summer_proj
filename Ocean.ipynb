```python
with open('data/bsi_dfo_surfacedrifters_2003.tab') as drifter:                                                                                          
    drifter_reader = csv.reader(drifter, delimiter='\t')
    drifters_data = []
    for drifter in drifter_reader:
#         print(drifter)
        if drifter[0].startswith('imlba03'):
            print(drifter)
            drifters_data.append(drifter)
        
#         for line in drifter:
#             print(line.strip(' '))
# rare = pd.read_csv('data/bsi_dfo_surfacedrifters_2003.tab',delimiter = '\t')
```
